```julia
mu, sigma = 100, 50

x = mu + sigma*rand(1000)

# the histogram of the data
n, bins, patches = plt.hist(x, 100, normed=1, alpha=0.75)

print("N:")
print(n)
print("\n\n bins:")
print(bins)
print("\n\n patches:")
print(patches)
```

```julia
mu, sigma = 100, 50

x = mu + sigma*randn(1000)

# the histogram of the data
n, bins, patches = plt.hist(x, 100, normed=1, facecolor="r",alpha=0.75)
```
